```python
import numpy as np
import pandas as pd

zu = np.arange(-20, 35, 0.1) + 273.15
ab = np.arange(0, 55, 0.1) +273.15

ZU, AB = np.meshgrid(zu, ab)
eta_max = AB / (AB - ZU) # maximaler Wirkungsgrad
eta_max = np.maximum(0, eta_max)
eta_max[eta_max == 0] = np.nan
eta_max[eta_max >= 30] = np.nan
```

```python
ny = 0.5 #Gütegrad
eta_real = eta_max * ny
```

```python
hp = pd.read_excel('data/ee_waermepumpen_anlagenliste_bis_2020.xlsx', header=[53, 54], decimal=',')

hp = hp[hp.columns[[0, 3, 27, 32, 36, 42, 47, 53, 57]]]

columns = ['Hersteller', 'Typ', 'A-7 / W35', 'A2 / W35', 'A7 / W35', 'A10 / W35', 'Wärme- Leistung A-7 / W35', 'Nennwärme- LeistungA2 / W35', 'Smart Grid Ready']
hp.columns = columns
hp = hp.dropna(subset=[hp.columns[0]])
hp = hp.dropna(subset=hp.columns[1:].values, how='all')
hp
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Surface(x = ZU-273.15, y = AB-273.15, z=eta_max, showscale=True, showlegend=True, cmin=0, cmax=30))#, surfacecolor=[0,30]))
fig.add_trace(go.Surface(x = ZU-273.15, y = AB-273.15, z=eta_real, showscale=True, showlegend=True, cmin=0, cmax=30))


fig.add_trace(go.Scatter3d(x = np.full(hp.shape[0], -7, dtype=np.int), y = np.full(hp.shape[0], 35, dtype=np.int), z = hp['A-7 / W35'], mode='markers', text=hp.Hersteller, hovertemplate= "<b>%{text}</b><br><br>" + "Typ: %{hp.Typ}" + "Efficiency: %{z:.0%}<br>" +"<extra></extra>"))
fig.add_trace(go.Scatter3d(x = np.full(hp.shape[0], 2, dtype=np.int), y = np.full(hp.shape[0], 35, dtype=np.int), z = hp['A2 / W35'], mode='markers', text=hp.Hersteller, hovertemplate= "<b>%{text}</b><br><br>" + "Typ: %{hp.Typ}" + "Efficiency: %{z:.0%}<br>" +"<extra></extra>"))
fig.add_trace(go.Scatter3d(x = np.full(hp.shape[0], 7, dtype=np.int), y = np.full(hp.shape[0], 35, dtype=np.int), z = hp['A7 / W35'], mode='markers', text=hp.Hersteller, hovertemplate= "<b>%{text}</b><br><br>" + "Typ: %{hp.Typ}" + "Efficiency: %{z:.0%}<br>" +"<extra></extra>"))
fig.add_trace(go.Scatter3d(x = np.full(hp.shape[0], 10, dtype=np.int), y = np.full(hp.shape[0], 35, dtype=np.int), z = hp['A10 / W35'], mode='markers', text=hp.Hersteller, hovertemplate= "<b>%{text}</b><br><br>" + "Typ: %{hp.Typ}" + "Efficiency: %{z:.0%}<br>" +"<extra></extra>"))
fig.update_layout(title='Max efficiency heat pump', autosize=True,
                  width=1500, height=1000,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.write_html('Heat Pump_' + 'plot.html', auto_open=False)
```

```python
# Wirtkungsgrad in Abh. von temperatur

eta = lambda ab, x: ab / (ab - x)
```

```python
import numpy as np
kelvin = lambda x: x + 273.15
eta_heating = lambda max_ab, min_ab, zu: np.log(((max_ab - zu)**max_ab*(min_ab - zu - 1)**(min_ab - zu)) / ((max_ab - zu - 1) ** (max_ab - zu) * (min_ab -zu) ** min_ab))
eta_heating(kelvin(35), kelvin(20), kelvin(2))
```

```python
def eta_heating(max_ab, min_ab, zu):
    x = np.linspace(min_ab, max_ab, (max_ab - min_ab) * 10 + 1)
    y = eta(kelvin(x), kelvin(zu))
    int_avg = trapz(y, x) / np.abs(x[-1] - x[0])

    return int_avg

```
